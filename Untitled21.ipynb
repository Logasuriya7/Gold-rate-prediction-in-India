```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
gold_dataset = pd.read_csv('/content/Gold vs USDINR.csv')
```

```python
gold_dataset.info()
```

**Data Analysis**

```python
gold_dataset.head()
```

```python
gold_dataset['Goldrate'] = gold_dataset['Goldrate'].replace('₹', '', regex = True).replace(',', '', regex = True).astype(float)
gold_dataset.head()
```

**Model Training**

```python
x = gold_dataset[['USD_INR']]
y = gold_dataset[['Goldrate']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1, random_state = 2)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
regresion = LinearRegression()
```

```python
regresion.fit(x_train_scaled, y_train)
```

```python
regresion.get_params()
```

```python
regresion.coef_
```

```python
regresion.intercept_
```

```python
m = regresion.coef_[0][0]
c = regresion.intercept_
```

```python
x_train_predict = regresion.predict(x_train_scaled)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, x_train_predict, color = 'red')
plt.xlabel('USD_INR')
plt.ylabel('Goldrate')

plt.show()
```

```python
x_test_predected = regresion.predict(x_test_scaled)
```

```python
x_test_predected
```

```python
y_test
```

```python
mean_squared_error(y_test, x_test_predected)
```

**Hyperperameter optimization**

```python
from sklearn.model_selection import RandomizedSearchCV
param_space = {'copy_X': [True, False],
               'fit_intercept': [True, False],
               'n_jobs': [1,2,10,15,None],
               'positive': [True, False]}
```

```python
search = RandomizedSearchCV(regresion, param_space, n_iter=50, cv=5)
```

```python
search.fit(x_train, y_train)
```

```python
search.best_params_
```

```python
tuned_model = LinearRegression(positive = False, n_jobs = 1, fit_intercept = True, copy_X = True)
```

```python
tuned_model.fit(x_train_scaled, y_train)
```

```python
tuned_model.coef_
```
